```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import ipywidgets as widgets
import csv
```

```python
df_pays = pd.read_csv("C:/Users/namor/Dropbox/Simplon/Programmation/Projet Fil rouge/Version 2/CSV/Pays.csv")
df_pays.insert(loc=0, column='Area_Code', value=df_pays['Area Code'])
df_pays = df_pays.set_index('Area_Code')


df_population = pd.read_csv("C:/Users/namor/Dropbox/Simplon/Programmation/Projet Fil rouge/Data/2018/FAOSTAT NB Populations.csv")
df_population = df_population.drop(columns=["Domain Code", "Domain", "Element Code", "Element", "Item Code", "Item", "Year Code", "Unit", "Flag", "Flag Description"])
df_population["Value"] = df_population["Value"] * 1000
df_population = df_population.rename(columns = {'Value': 'Populations'})
df_population.insert(loc=0, column='Area_Code', value=df_population['Area Code'])
df_population = df_population.set_index('Area_Code')


df_nutrition = pd.read_csv("C:/Users/namor/Dropbox/Simplon/Programmation/Projet Fil rouge/Data/2018/FAOSTAT NB Populations en sous-nutrition.csv")
df_nutrition = df_nutrition.drop(columns=["Domain Code", "Domain", "Element Code", "Element", "Item Code", "Item", "Year Code", "Unit", "Flag", "Flag Description", "Note"])
index_with_nan = df_nutrition.index[df_nutrition.isnull().any(axis=1)]
df_nutrition.drop(index_with_nan, 0, inplace=True)
df_nutrition['Value'] = df_nutrition['Value'].apply(lambda x: str(x) + " " + "M")
df_nutrition['Year'] = df_nutrition['Year'].apply(lambda x: 2018)
df_nutrition = df_nutrition.rename(columns = {'Value': 'Sous-Nutrition'})
df_nutrition.insert(loc=0, column='Area_Code', value=df_nutrition['Area Code'])
df_nutrition = df_nutrition.set_index('Area_Code')


df_capital = pd.read_csv("C:/Users/namor/Dropbox/Simplon/Programmation/Projet Fil rouge/Data/2018/FAOSTAT Investissement, Dépréciation, Stock Financier.csv")
df_capital = df_capital.drop(columns=["Domain Code", "Domain", "Element Code", "Element", "Item Code", "Item", "Year Code", "Year", "Unit", "Flag", "Flag Description", "Note"])
df_capital.drop_duplicates(subset = "Area", keep = 'first', inplace=True)
df_capital = df_capital.rename(columns = {'Value': 'Investissement Alimentaire'})
df_capital[['Investissement Alimentaire']] = df_capital[['Investissement Alimentaire']].astype(np.int64)
df_capital['Investissement Alimentaire'] = df_capital['Investissement Alimentaire'].round().apply(lambda x: str(x) + " " + "M")
df_capital.insert(loc=0, column='Area_Code', value=df_capital['Area Code'])
df_capital = df_capital.set_index('Area_Code')

df_sante = pd.read_csv("C:/Users/namor/Dropbox/Simplon/Programmation/Projet Fil rouge/Version 2/CSV/Santé.csv")
df_sante = df_sante[df_sante["Year"] == 2019]
df_sante.insert(loc=0, column='Area_Code', value=df_sante['Area Code'])
df_sante = df_sante.set_index('Area_Code')

liste = []
for i in range(255):
    liste.append(2018)

df_details_pays = pd.merge(df_pays, df_population["Populations"], how='left', left_index=True, right_index=True)
df_details_pays = pd.merge(df_details_pays, df_nutrition["Sous-Nutrition"], how='left', left_index=True, right_index=True)
df_details_pays = pd.merge(df_details_pays, df_capital["Investissement Alimentaire"], how='left', left_index=True, right_index=True)
df_details_pays = pd.merge(df_details_pays, df_sante[["Life expectancy at birth (Both)", "Life expectancy at age 60 (Both)", "Healthy life expectancy at birth (Both)", "Healthy life expectancy at age 60 (Both)"]], how='left', left_index=True, right_index=True)
df_details_pays = df_details_pays.reset_index(drop=True)
del df_details_pays['Area']
df_details_pays.insert(loc=1, column='Year', value=liste)
df_details_pays.drop_duplicates(subset = "Area Code", keep = 'first', inplace=True)

# Enregistrer le DataFrame dans un fichier csv
# df_details_pays.to_csv('CSV/Details Pays.csv', index=False, encoding="utf-8-sig", quoting=csv.QUOTE_NONNUMERIC)
df_details_pays.to_csv('CSV/Details Pays.csv', index=False, sep=',', encoding='utf-8-sig')
df_details_pays
```
