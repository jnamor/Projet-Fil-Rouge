```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import ipywidgets as widgets
import csv
```

```python
df_terres = pd.read_csv("C:/Users/namor/Dropbox/Simplon/Programmation/Projet Fil rouge/Data/2018/FAOSTAT Type de terres.csv")

# Supprimer les colonnes inutiles
df_terres = df_terres[['Area', 'Year', 'Item', 'Value']]
df_terres[['Value']] = df_terres[['Value']].astype(np.int64)
df_terres['Value'] = df_terres['Value'].apply(lambda x: str(x) + " " + "%")

df_terres.insert(loc=0, column='Area_1', value=df_terres['Area'])
df_terres = df_terres.set_index('Area_1')

df_pays = pd.read_csv("C:/Users/namor/Dropbox/Simplon/Programmation/Projet Fil rouge/Version 2/CSV/Pays.csv")
df_pays.insert(loc=0, column='Area_1', value=df_pays['Area'])
df_pays = df_pays.set_index('Area_1')

df_terres_7 = df_terres[df_terres['Item'].str.contains('Arable land')]
df_terres_7 = df_terres_7.rename(columns={"Value":'Arable land'})

df_terres_1 = df_terres[df_terres['Item'].str.contains('Land under permanent crops')]
df_terres_1 = df_terres_1.rename(columns={"Value":'Land under permanent crops'})

df_terres_2 = df_terres[df_terres['Item'].str.contains('Cropland')]
df_terres_2 = df_terres_2.rename(columns={"Value":'Cropland'})

df_terres_3 = df_terres[df_terres['Item'].str.contains('Land under perm. meadows and pastures')]
df_terres_3 = df_terres_3.rename(columns={"Value":'Land under perm. meadows and pastures'})

df_terres_4 = df_terres[df_terres['Item'].str.contains('Agriculture area actually irrigated')]
df_terres_4 = df_terres_4.rename(columns={"Value":'Agriculture area actually irrigated'})

df_terres_5 = df_terres[df_terres['Item'].str.contains('Land area equipped for irrigation')]
df_terres_5 = df_terres_5.rename(columns={"Value":'Land area equipped for irrigation'})

df_terres_6 = df_terres[df_terres['Item'].str.contains('Agriculture area under organic agric.')]
df_terres_6 = df_terres_6.rename(columns={"Value":'Agriculture area under organic agric.'})


df_entite_terres = pd.merge(df_pays, df_terres_7["Arable land"], how='left', left_index=True, right_index=True)
df_entite_terres = pd.merge(df_entite_terres, df_terres_1["Land under permanent crops"], how='left', left_index=True, right_index=True)
df_entite_terres = pd.merge(df_entite_terres, df_terres_2["Cropland"], how='left', left_index=True, right_index=True)
df_entite_terres = pd.merge(df_entite_terres, df_terres_3["Land under perm. meadows and pastures"], how='left', left_index=True, right_index=True)
df_entite_terres = pd.merge(df_entite_terres, df_terres_4["Agriculture area actually irrigated"], how='left', left_index=True, right_index=True)
df_entite_terres = pd.merge(df_entite_terres, df_terres_5["Land area equipped for irrigation"], how='left', left_index=True, right_index=True)
df_entite_terres = pd.merge(df_entite_terres, df_terres_6["Agriculture area under organic agric."], how='left', left_index=True, right_index=True)

liste = []

for i in range(255):
    liste.append(2018)
    
df_entite_terres = pd.merge(df_entite_terres, df_pays)
del df_entite_terres['Area']
df_entite_terres.insert(1, column= "Year", value=liste)
# df_entite_terres = df_entite_terres[['Area Code', 'Year', 'Arable land', "Land under permanent crops", "Cropland", "Land under perm. meadows and pastures", "Agriculture area actually irrigated", "Land area equipped for irrigation", "Agriculture area under organic agric."]]
df_entite_terres.insert(loc=0, column='Area_Code', value=df_entite_terres['Area Code'])
df_entite_terres = df_entite_terres.set_index('Area_Code')

df_meteo = pd.read_csv("C:/Users/namor/Dropbox/Simplon/Programmation/Projet Fil rouge/Version 2/CSV/MÃ©tÃ©o.csv")
df_meteo = df_meteo[df_meteo['Year'] == 2016]
df_meteo.insert(loc=0, column='Area_Code', value=df_meteo['Area Code'])
df_meteo = df_meteo.set_index('Area_Code')

# df_entite_terres = pd.merge(df_entite_terres, df_meteo[["Temperature (Â°C)", "Rainfall - (MM)"]], how='left', left_index=True, right_index=True)
df_entite_terres = df_entite_terres.reset_index(drop=True)

# Enregistrer le DataFrame dans un fichier csv
df_entite_terres.drop_duplicates(subset = "Area Code", keep = 'first', inplace=True)
# df_entite_terres.to_csv('CSV/Terres.csv', index=False, encoding="utf-8-sig", quoting=csv.QUOTE_NONNUMERIC)
df_entite_terres.to_csv('CSV/Terres.csv', index=False, sep=',', encoding="utf-8-sig")
df_entite_terres
```
